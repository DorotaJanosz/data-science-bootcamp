```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
fir = plt.figure()
plt.show()
```

###Wykres liniowy

```
plt.plot([0,1],[0,3], label='line', color='green')
plt.legend()
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.title('Wykres')
plt.show()
```

```
plt.plot([0,1,2],[0,3,4], label='line', color='green')
plt.legend()
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.title('Wykres')
plt.show()
```

```
x = np.arange(-3, 3, 0.1)
sin_x = np.sin(x)
cos_x = np.cos(x)
plt.plot(x, sin_x)
_ = plt.plot(x, cos_x)
```

###Tworzenie wielu wykresów

```
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
```

```
fig = plt.figure(figsize=(12,8))
plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 1, 4], 'ro')

fig.show()
```

###Wykresy słupkowe

```
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
```

```
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
```

```
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])
```

###Wykresy punktowe

```
x1 = np.random.randn(300)
x2 = np.random.randn(300)
size = np.random.randn(300)*50

plt.scatter(x=x1, y=x2, s=size, alpha=0.5)
```
