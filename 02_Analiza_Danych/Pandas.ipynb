###Data Frame w porwnaiu do jzyka SQl


```
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
df_raw = pd.read_excel(url)
df_raw.head()
```

```
df = df_raw.copy()
```

```
df.info()
```

```
df.isnull().sum()
```

```
df = df.dropna()
df.isnull().sum()
```

```
df.describe()
```

```
df = df[df.Quantity >= 0]
df.info()
```

```
df.CustomerID = df.CustomerID.apply(lambda x: str(int(x)))
```

```
# Select * from df
df
```

```
# Select Quantity, UnitPrice, CustomerID drom df;
df[['Quantity', 'UnitPrice', 'CustomerID']]
```

```
# Select Quantity, UnitPrice, CustomerID drom df limit 10;
df[['Quantity', 'UnitPrice', 'CustomerID']][:10]
```

```
# Select * from df where CustomerID ='17850';
df.CustomerID == '17850'
df[df.CustomerID == '17850']
df.query('CustomerID == "17850"')
```

```
# Select * from df where CustomerID ='17850' and UnitPrice >5;
df[(df.CustomerID == '17850') & (df.UnitPrice > 5)]
```

```
# Select * from df where CustomerID ='17850' or Country = "France";
df[(df.CustomerID == '17850') | (df.Country == 'France')]
```

```
# Select * from df where InvoiceNu is not null;
df[df.InvoiceNo.notnull()]
```

```
# Select CustomerID, count(*) from df group by CustomerID;
df.groupby(by='CustomerID').size()
```

```
df["Revenue"] = df.Quantity * df.UnitPrice
df
```

```
# Select CustomerID, avg(Revenue), count(*) from df group by CustomerID;
import numpy as np

df.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size})
df.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size}).rename(columns={'Revenue': 'Revenue_Average', 'CustomerID': 'CustomerID_Count'})
```

```
df.groupby('CustomerID')['Revenue', 'CustomerID'].agg(['mean', 'size'])
```

```
df['InvoiceDateDay'] = df.InvoiceDate.dt.day
```

```
# Select InvoiceDateDay, sum(Revenue) from df group by InvoiceDateDay

results = df.groupby('InvoiceDateDay').aggregate({'Revenue': np.sum})
```

```
import seaborn as sns
sns.set()

results.plot(kind='bar')
```
